```python
import cv2, os, json
import numpy as np
import mediapipe as mp

mp_pose = mp.solutions.pose

for data_dir in os.listdir("dataset"):
  print("Current folder: " + data_dir)
  for filename in os.listdir('dataset/' + data_dir):
      image = cv2.imread('dataset/' + data_dir + '/' + filename)
      try:
        with mp_pose.Pose(
        static_image_mode=True, min_detection_confidence=0.8) as pose:

          results = pose.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        if not results.pose_landmarks:
          print("No human detected: " + filename)
          continue
        else:
          landmarks = results.pose_landmarks.ListFields()[0][1]
          landmarks_export = []
          for landmark in landmarks:
            landmarks_export.append(landmark.x)
            landmarks_export.append(landmark.y)
            landmarks_export.append(landmark.z)
          with open(
              'gdrive/MyDrive/fall-detection-m4/mediapipe_output/' + data_dir + '/'
              + filename + ".json", "w") as landmarks_output:
            landmarks_output.write(str(landmarks_export))
      except:
        print("Error with the following file: " + filename)
```

```python
def extend_header(header):
  extended_header = []
  for keypoint in header:
    extended_header.append(keypoint + "_x")
    extended_header.append(keypoint + "_y")
    extended_header.append(keypoint + "_confidence")
  return extended_header
```

```python
def create_df(folder_nbr):
  df = pd.DataFrame([], columns=extend_header(header))
  index = 0
  for filename in os.listdir('gdrive/MyDrive/fall-detection-m4/openpifpaf_output/openpifpaf_output/' + folder_nbr):
    with open('gdrive/MyDrive/fall-detection-m4/openpifpaf_output/openpifpaf_output/' 
              + folder_nbr + '/' + filename) as json_sample:
      sample = json.load(json_sample)
      try:
        row = sample[0]["keypoints"]
        df.loc[index] = row
        index += 1
      except:
        print("This file might be compromised: " + filename)
  df.to_pickle("gdrive/MyDrive/fall-detection-m4/openpifpaf_output/openpifpaf_output/"+ folder_nbr + ".pkl")
  return True
```
